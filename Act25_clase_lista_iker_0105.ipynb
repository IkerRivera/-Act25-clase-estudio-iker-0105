```python
# Definición de la Clase Asistencia
class Asistencia:
    def __init__(self):
        # Inicializamos los atributos vacíos
        self.nombre = ""
        self.fecha = ""
        self.estado = ""

    # Función 1: Leer datos desde el teclado
    def leerdatos(self):
        print("\n--- Registro de Asistencia ---")
        self.nombre = input("Ingrese nombre del alumno: ")
        self.fecha = input("Ingrese la fecha (dd/mm/aaaa): ")
        self.estado = input("Ingrese el estado (Asistió/Faltó/Retraso): ")

    # Función 2: Mostrar los datos guardados
    def mostrardatos(self):
        print("\n--- Ficha de Asistencia ---")
        print(f"Alumno: {self.nombre}")
        print(f"Fecha:  {self.fecha}")
        print(f"Estado: {self.estado}")
        print("-" * 25)

# --- Creación de Objetos ---

# Creamos el primer objeto
asistencia1 = Asistencia()
asistencia1.leerdatos()

# Creamos el segundo objeto
asistencia2 = Asistencia()
asistencia2.leerdatos()

# --- Mostrar los resultados ---
print("\nRESUMEN FINAL:")
asistencia1.mostrardatos()
asistencia2.mostrardatos()
print("programa realizado por Iker Montoya")
```
