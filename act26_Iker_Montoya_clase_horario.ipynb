```python
# Definición de la Clase Horario
class Horario:
    def __init__(self):
        # Inicializamos los 3 atributos
        self.dia = ""
        self.hora_inicio = ""
        self.hora_fin = ""

    # Función 1: leerdatos() - Interactiva con el usuario
    def leerdatos(self):
        print("\n--- Ingreso de datos del Horario ---")
        self.dia = input("Ingrese el día (ej. Lunes): ")
        self.hora_inicio = input("Ingrese la hora de inicio (ej. 08:00): ")
        self.hora_fin = input("Ingrese la hora de fin (ej. 10:00): ")

    # Función 2: mostrardatos() - Imprime la información
    def mostrardatos(self):
        print("\n--- Información del Horario ---")
        print(f"Día establecido: {self.dia}")
        print(f"Rango de tiempo: de {self.hora_inicio} a {self.hora_fin}")
        print("-" * 30)

# --- Creación de los 2 Objetos ---

# Objeto 1
horario1 = Horario()
horario1.leerdatos()

# Objeto 2
horario2 = Horario()
horario2.leerdatos()

# --- Mostrar los resultados finales ---
print("\nREGISTROS CREADOS:")
horario1.mostrardatos()
horario2.mostrardatos()
```
